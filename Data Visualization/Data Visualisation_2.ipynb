```python
import csv

def convert_time_to_num(time):
  mins = int(time[-2:])
  frac_of_hour = mins/60.0
  hour = int(time[:-3])
  time = hour + frac_of_hour
  return time

sales_times_raw = []
with open('/Users/elorm/Documents/Repos/Datasets/sales_times.csv') as csvDataFile:
  csvReader = csv.reader(csvDataFile)
  for row in csvReader:
    sales_times_raw.append(row[2])
  sales_times_raw = sales_times_raw[1:]

sales_times = []
for time in sales_times_raw:
  sales_times.append(convert_time_to_num(time))
```

```python
#create the histogram here
plt.hist(sales_times)
plt.show()
```

```python
#create the histogram here
plt.hist(sales_times, bins = 20)
plt.show()
```

#### Orders Over Time
FoodWheel is a relatively new startup. They launched in April, and have grown more popular since then. Management suspects that the average order size has increased over time. They’d like you to investigate if this claim is true and answer these questions:

- How has the average order amount changed over time?
- What does this say about the trajectory of the company?

```python
orders = pd.read_csv('/Users/elorm/Documents/Repos/Datasets/orders.csv')
print(orders.head())
orders.info()
```

```python
#Creating a new column called month
orders['month'] = orders['date'].str.split('-')
orders['month'] = orders.apply(lambda row: row['month'][0], axis = 1)
orders.head()
```

```python
#Finding average monthly purchases
avg_order = orders.groupby('month')['price'].agg('mean').reset_index()
print(avg_order)
```

```python
std_order = orders.groupby('month').price.std().reset_index()
print(std_order)
```

#### Orders Over Time

```python
ax = plt.subplot()
bar_heights = avg_order['price']
bar_errors = std_order['price']
plt.bar(range(len(avg_order['price'])), bar_heights, yerr = bar_errors, capsize = 5)
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['April', 'May', 'June', 'July', 'August', 'September'])
plt.title('Average Sales')
plt.ylabel('Average Price')
plt.xlabel('Month')
plt.show()
```
